```python
!{sys.executable} -m pip install --user numpy
import numpy as np
!{sys.executable} -m pip install --user pandas
import pandas as pd
!{sys.executable} -m pip install --user seadorn
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Perceptron
```

```python
data = pd.read_csv('isolet.csv')
data
```

```python
X_data = data.iloc[: , :617]
Y_data = data.iloc[: , 617:]
Y_data

```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, Y_data, test_size=0.2, random_state=42)
X_train
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
X_test
```

```python
clf = MLPClassifier(hidden_layer_sizes=20,learning_rate='adaptive',alpha=0.1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Точность модели: {accuracy}')
```

```python
p = Perceptron(penalty='l1', alpha=0.0001)
p.fit(X_train, y_train)
y_pred = p.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Точность модели Perceptron: {accuracy}')
```
